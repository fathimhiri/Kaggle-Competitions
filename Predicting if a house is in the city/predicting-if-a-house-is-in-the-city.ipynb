```python
y=data['city']
x=data.drop('city', axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,train_size=0.8)
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
```

```python
log_model= LogisticRegression(random_state=0,verbose =1)
svm_model= svm.SVC(kernel='rbf',verbose =1)

nn_model= MLPClassifier(random_state=1, max_iter=300,verbose =1)
```

```python
log_model.fit(x_train, y_train)
svm_model.fit(x_train, y_train)
nn_model.fit(x_train, y_train)
```

```python
print(log_model.score(x_train, y_train))
print(svm_model.score(x_train, y_train))
print(nn_model.score(x_train, y_train))
```

```python
from sklearn.metrics import f1_score
```

```python
log_pred = log_model.predict(x_test)
svm_pred = svm_model.predict(x_test)
nn_pred = nn_model.predict(x_test)
```

```python
print(f1_score(log_pred, y_test))
print(f1_score(svm_pred, y_test))
print(f1_score(nn_pred, y_test))
```
